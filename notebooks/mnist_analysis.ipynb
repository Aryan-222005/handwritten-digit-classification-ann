### 🧠 1. Problem Statement

#### Handwritten Digit Classification using ANN

The goal of this project is to build an Artificial Neural Network (ANN) model 
that can classify handwritten digits (0–9) from the MNIST dataset.

This is a multi-class classification problem where each image belongs to one of 10 classes.

### 📊 2. Dataset Overview

```python
from tensorflow.keras.datasets import mnist

(X_train, y_train), (X_test, y_test) = mnist.load_data()

print("Training shape:", X_train.shape)
print("Testing shape:", X_test.shape)
```

- Dataset contains grayscale images of size 28x28
- Total samples: 70,000
- Each pixel value ranges from 0 to 255

### 3. Data Visualization

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

for i in range(10):
    plt.subplot(2,5,i+1)
    plt.imshow(X_train[i], cmap='gray')
    plt.title(f"Label: {y_train[i]}")
    plt.axis('off')

plt.show()
```

Observation:
- Digits vary in writing style
- Some digits are visually similar → potential misclassification

### ⚙️ 4. Data Preprocessing

```python
# Normalize
X_train = X_train / 255.0
X_test = X_test / 255.0

# Flatten
X_train = X_train.reshape(-1, 784)
X_test = X_test.reshape(-1, 784)

from tensorflow.keras.utils import to_categorical

y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

Why preprocessing is needed:
- Normalization → faster convergence
- Flattening → ANN requires 1D input
- One-hot encoding → multi-class classification

### 🧠 5. Model Architecture (Concept)
The ANN consists of:
- Input layer: 784 neurons (flattened image)
- Hidden layers: Dense layers with ReLU activation
- Output layer: 10 neurons with Softmax

ReLU is used to introduce non-linearity.
Softmax converts outputs into probabilities.

### ❌ 8. Error Analysis

```python
Image.open("outputs/wrong_predictions.png")
```

Observations:

Misclassifications occur mainly due to:

- Similar digit structures (e.g., 8 vs 3, 9)
- Variations in handwriting styles
- Loss of spatial information due to flattening

For example, digit '8' has multiple loops and is sometimes misclassified 
as '3' or '9', leading to slightly lower recall.

### 🧠 9. Key Learnings

ANN does not capture spatial relationships because the image is flattened 
into a 1D vector. This removes positional information of pixels.

As a result, ANN cannot effectively understand shapes and patterns 
compared to CNN, which preserves spatial structure.

Possible improvements:
